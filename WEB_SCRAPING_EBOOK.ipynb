# EBOOK SCRAPING

```python
import os
import time
import urllib.request
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from openpyxl import *
import openpyxl

input_book = input("Search book: ")

wb = Workbook()

ws = wb.active

ws.title = "Books"

title_pos = ws.cell(1,1)
title_pos.value = "Book name"

publisher_pos = ws.cell(1,2)
publisher_pos.value = "Publisher"

author_pos = ws.cell(1,3)
author_pos.value = "Author"

p_year_pos = ws.cell(1,4)
p_year_pos.value = "Published year"

language_pos = ws.cell(1,5)
language_pos.value = "Language"

format_pos = ws.cell(1,6)
format_pos.value = "File format"

image_pos = ws.cell(1,7)
image_pos.value = "Image"

driver = webdriver.Chrome("chromedriver")
url = 'https://1lib.eu/'
driver.get(url)

time.sleep(2)

search_input = driver.find_element_by_id('searchFieldx')
search_input.send_keys(input_book)

time.sleep(2)

search_buton = driver.find_element_by_class_name('inner')
search_buton.click()

pages = driver.find_elements_by_css_selector('div.paginator span a')

num_books = 0

pages = driver.find_elements_by_css_selector('div.paginator span a')

num_books = 0

num_page = 0

count_line = 2

count_img = 1

for page in range(len(pages)+1):
    
    time.sleep(3)
    
    count = 0

    each_book = driver.find_elements_by_css_selector('table.resItemTable')

    for book in each_book:

        b_name = driver.find_elements_by_tag_name('h3 a')[count].text

        try:

            find_pub = driver.find_elements_by_css_selector('table.resItemTable')[count]

            b_publisher = find_pub.find_element_by_css_selector("[title^='Publisher'] a").text

        except:
            b_publisher = "Error publisher"


        b_author = driver.find_elements_by_css_selector('div.authors')[count].text

        find_details = driver.find_elements_by_css_selector('div.bookDetailsBox')[count]
        
        try:
            b_year = find_details.find_element_by_css_selector('div.property_year div.property_value').text

        except:
            b_year = "Error year"

       
        try:

            b_lang = find_details.find_element_by_css_selector('div.property_language div.property_value').text

        except:
            b_lang = "Error language"
      
        
        try:

            b_format_with_coma= find_details.find_element_by_css_selector('div.property__file div.property_value').text
            b_format = b_format_with_coma.split(',', 1)[0]

        except:
            b_format = "Error format"
            

        print(b_name)

        print(b_publisher)

        print(b_author)

        print(b_year)
        print(b_lang)
        print(b_format)

        try:    
            b_cover= driver.find_elements_by_css_selector('img.cover')[count]
            b_cover_src = b_cover.get_attribute("data-src")

            if not b_cover_src:
                b_cover_src = b_cover.get_attribute("src")

            elif b_cover_src == "/img/cover-not-exists.png" or b_cover_src=="https://1lib.eu/img/book-no-cover.png":
                b_cover_src = "Imagen no disponible"
        except:
            b_cover_src = "Imagen no disponible"

        print(b_cover_src)

        print("\n")
        
        insert_title = ws.cell(count_line,1)
        insert_title.value = b_name

        insert_publisher = ws.cell(count_line,2)
        insert_publisher.value = b_publisher

        insert_author = ws.cell(count_line,3)
        insert_author.value = b_author

        insert_year = ws.cell(count_line,4)
        insert_year.value = b_year

        insert_language = ws.cell(count_line,5)
        insert_language.value = b_lang

        insert_format = ws.cell(count_line,6)
        insert_format.value = b_format

        insert_image = ws.cell(count_line,7)
        
        if not (b_cover_src == "Imagen no disponible"):
            length = len(b_cover_src)
            img_extension = b_cover_src[length - 4 :]
            name_img =  str(count_img) + img_extension
            urllib.request.urlretrieve(b_cover_src, name_img)
            img = openpyxl.drawing.image.Image(name_img)
            img.anchor = 'G' + str(count_line)
            ws.add_image(img)

            count_img += 1
            
        else:
            insert_image.value = b_cover_src
            
        count += 1
        
        count_img += 1
        
        num_books += 1
        
        count_line += 1
        
    time.sleep(5)
    
    try:
        next_page = driver.find_elements_by_css_selector('div.paginator span a')[num_page]
        num_page+=1
        next_page.click()
    
    except:
        print("No hay más resultados")
    

wb_name_pref = input_book.replace(" ","_")

wb_name = wb_name_pref+".xlsx"

wb.save(filename = wb_name)
    
print("Numero de libros encontrados:",num_books)
print("Programa terminado")

driver.quit()
```
